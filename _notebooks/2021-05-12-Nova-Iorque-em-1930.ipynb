```python
#hide
from deoldify_wrapper import DeOldify
```

```python
#hide
deo = DeOldify()
```

```python
#hide_input
source_url = 'https://nypost.com/wp-content/uploads/sites/2/2016/08/union-square.jpg?quality=90&strip=all&w=978&h=652&crop=1' #@param {type:"string"}
render_factor = 30  #@param {type: "slider", min: 7, max: 40}
compare = True #@param {type: "boolean"}

if source_url is not None and source_url !='':
    output = deo.colorize(filepath_or_url=source_url, render_factor=render_factor, compare=compare)
else:
    print('Provide an image url and try again.')
```

A colorização foi muito bem feita, é possível ver a propaganda dos cigarros Camel com detalhes da embalagem, assim como o *outdoor* da cerveja Ruppert. Alguns pequenos defeitos foram o azulado no rosto das pessoas e os *outdoors* laterais continuaram em preto e branco, no geral foi uma ótima modificação.
